```python
# 참조하기

import pandas as pd
import requests
from urllib.request import urlopen
import urllib
from bs4 import BeautifulSoup


target_url = "http://www.koreabaseball.com/Record/Team/Hitter/BasicOld.aspx"

translated_name = {"두산": "Dusan",
                   "넥센": "Nexen",
                   "KIA": "KIA",
                   "SK": "SK",
                   "한화": "Hangwa",
                   "LG": "LG",
                   "NC": "NC",
                   "롯데": "Lotte",
                   "삼성": "Samsung",
                   "KT": "KT"}


class TeamInformation:
    """
    The general information of teams in KBO.
    """

    def __init__(self):
        self._URL = "http://www.koreabaseball.com/Record/Team/Hitter/BasicOld.aspx"
        self.translated_name = {"두산": "Dusan",
                           "넥센": "Nexen",
                           "KIA": "KIA",
                           "SK": "SK",
                           "한화": "Hangwa",
                           "LG": "LG",
                           "NC": "NC",
                           "롯데": "Lotte",
                           "삼성": "Samsung",
                           "KT": "KT"}

    def hit_stat(self):
        """
        Return the hitting statistics of KBO teams.
        :return:
        """
        data = {}
        source_code = requests.get(self._URL)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text, "html.parser")
        team_list = soup.find_all("tr")
        # Clean the text.
        for i in range(len(team_list)):
            if team_list[i].text.replace("<td>", ""):
                stat_list = tuple(team_list[i + 1 : i + 17])
                data[team_list[i].text.replace("<td>", "").replace("</td>", "")] = stat_list
        return data

if __name__ == "__main__":
    rd = TeamInformation()
    print(rd.hit_stat())
```
